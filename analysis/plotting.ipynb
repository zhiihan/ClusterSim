```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("./data/2025_05_20.csv").sort_values(by=['unit_cube_scale', 'p'])

averaged_df = df.groupby(['unit_cube_scale', 'p']).mean().reset_index()
df = averaged_df

shape_x = int(df.iloc[0]['shape_x'])
shape_y = int(df.iloc[0]['shape_y'])
shape_z = int(df.iloc[0]['shape_z'])

sample_size = int(df['sample'].max()) + 1

plt.rcParams.update({
    "text.usetex": True,
    'font.family': 'serif',
    'font.serif': ['Computer Modern']    
    })

df['unit_cells_ratio'] = df['unit_cells']/df['unit_cells'].max()

#plt.figure(figsize=(10, 6))
for scale in df['unit_cube_scale'].unique():
    subset = df[df['unit_cube_scale'] == scale]
    plt.plot(subset['p'], subset['unit_cells'], marker='o', label=f'Scale factor = {scale}')


plt.title(f"$p$ vs. Number of unit cells found, shape = ({shape_x}, {shape_y}, {shape_z}), samples = {sample_size}")
plt.xlabel("Probability of deletion $p$")
plt.ylabel("Number of unit cells found")
plt.legend(title="Scale")
plt.savefig("data/unit_cells.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
